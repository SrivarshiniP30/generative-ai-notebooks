```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("data/chase_banking.pdf")
documents= loader.load()
print(documents)
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator="",
    chunk_size= 10,
    chunk_overlap=2,
)
```

```python
chunks=[]

for doc in documents:
    texts = text_splitter.split_text(doc.page_content)
    chunks.extend(texts)
```

```python
for chunk in enumerate(chunks):
    print(chunk)
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain.embeddings import HuggingFaceEmbeddings

# Step 1: Set up the embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Step 2: Set up the semantic chunker
chunker = SemanticChunker(embedding_model, breakpoint_threshold_type="percentile", breakpoint_threshold_amount=95)

# Step 3: Input text
long_text = """
LangChain is a framework for developing applications powered by language models.
It enables applications that are data-aware, agentic, and composable.
LangChain supports integrations with many models and tools and provides abstraction for chains, agents, and more.
With LangChain, developers can easily connect LLMs to external data sources and create reasoning workflows.
"""

# Step 4: Chunk the text semantically
chunks = chunker.split_text(long_text)

# Step 5: View the output
for i, chunk in enumerate(chunks, 1):
    print(f"--- Chunk {i} ---")
    print(chunk)
```
